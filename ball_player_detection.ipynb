```python
import cv2
import numpy as np

# Lire l'image en niveaux de gris
gray_image = cv2.imread('output_videos/Capture.PNG', cv2.IMREAD_GRAYSCALE)

# Détecter les points d'intérêt
corners = cv2.goodFeaturesToTrack(gray_image, maxCorners=100, qualityLevel=0.01, minDistance=10)

# Afficher les coins détectés sur l'image
for corner in corners:
    x, y = corner.ravel()
    x, y = int(x), int(y)  # Conversion en entiers
    cv2.circle(gray_image, (x, y), 3, 255, -1)

cv2.imshow('Corners', gray_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
